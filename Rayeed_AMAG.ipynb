# **For T1 and T2**

```python
T1_new=T1.merge(T2, on='Time (s)')
```

```python
T1_new
```

```python
T1_new.columns
```

```python
dist1_2=[]
m = T1_new.shape[0]
for i in range (0,m):
  dist1_2.append(hs.haversine((T1_new['Latitude_y'][i],T1_new['Longitude_y'][i]),(T1_new['Latitude_x'][i],T1_new['Longitude_x'][i]))*1000)
```

```python
dist1_2
```

As they are moving in the opposite direction, the distance between them is increasing.

# **T1 and T2_2**

```python
T1_2new=T1.merge(T2_2, on='Time (s)')
```

```python
T1_2new
```

```python
dist1_2_2=[]
m = T1_2new.shape[0]
for i in range (0,m):
  dist1_2_2.append(hs.haversine((T1_2new['Latitude_y'][i],T1_2new['Longitude_y'][i]),(T1_2new['Latitude_x'][i],T1_2new['Longitude_x'][i]))*1000)
```

```python
dist1_2_2
```

```python
T1_2new['xl-xf'] = dist1_2_2
```

```python
T1_2new['xl-xf-d'] = T1_2new['xl-xf'] - 3 
```

```python
speed_dif=[]
speed_dif.append(0)
for i in range (1,m):
  speed_dif.append((T1_2new['xl-xf'][i-1] - T1_2new['xl-xf'][i])/(T1_2new['Time (s)'][i] - T1_2new['Time (s)'][i-1]))

speed_dif
```

```python
T1_2new['vf-vi']=speed_dif
```

```python
T1_2new
```

```python
TTC=[]
for i in range(0,m):
  if float(T1_2new['vf-vi'][i])>0:
    TTC.append(T1_2new['xl-xf-d'][i]/T1_2new['vf-vi'][i])
  else:
    TTC.append('NA')

TTC
```

```python
T1_2new['TTC']=TTC
T1_2new
```

```python
T1_2new.to_csv('/content/TTC_1_2_2.csv')
```
